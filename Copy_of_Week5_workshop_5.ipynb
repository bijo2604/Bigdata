## **Step 3: Data Cleaning**

```python
#Handle missing values:
#Drop rows (if needed):
df_cleaned = df.dropna()
```

```python
#Fill missing values using imputation:
df.fillna(df.mode(), inplace=True)  # Mean imputation for numerical columns
df.fillna(df.mode().iloc[0], inplace=True)  # Mode imputation for categorical columns
```

```python
#Check for duplicates and remove if necessary:
df.duplicated().sum()  # Count duplicates
df = df.drop_duplicates()  # Remove duplicates
```

## **Step 6: Identify Outliers**

### **link for understanding boxplot:**

*  https://www.simplypsychology.org/boxplots.html

* https://builtin.com/data-science/boxplot

* https://datatab.net/tutorial/box-plot

```python
#Use boxplots to detect outliers:
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.show()
```

```python
#Use the IQR (Interquartile Range) method to remove outliers:
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```
